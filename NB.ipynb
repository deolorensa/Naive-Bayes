Preprocessing

```python
#Menghapus kolom id
df.drop(columns={'ID'}, inplace=True)
df.head()
```

```python
#Rename kolom
df.columns=[column.lower() for column in df.columns]
df.columns
```

```python
#Validasi NaN/null
print(df.isna().sum())
```

```python
#Delete duplicated data
df=df.drop_duplicates()
df.head()
```

```python
#Convert stringg value to Int
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
df.iloc[:, 1:]=encoder.fit_transform(df.iloc[:,1:])
df.head
```

```python
#Memisah data menjadi feature dan class
x=df.iloc[:,1:]
y=df[{'class'}]
```

```python
# Tahap Preprocessing 7
# Memisah data train dan validasi
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, random_state = 1)
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_validate,cross_val_score
model=GaussianNB()
X=df.drop('class',axis=1)
y=df['class']
cv_score1=cross_validate(model,X,y,cv=10, return_train_score=True)
cv_score2=cross_val_score(model,X,y,cv=10)
```

```python
print (cv_score1['train_score'].mean(), cv_score1['test_score'].mean())
```

```python
print(cv_score2.mean()) #test_score
```

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import cross_validate,cross_val_score
model=CategoricalNB()
X=df.drop('class',axis=1)
y=df['class']
cv_score1=cross_validate(model,X,y,cv=10, return_train_score=True)
cv_score2=cross_val_score(model,X,y,cv=10)
```

```python
print (cv_score1['train_score'].mean(), cv_score1['test_score'].mean())
```

```python
print(cv_score2.mean()) #test_score
```

Evaluate Data

```python
#Turning hyperparameter
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=2, shuffle=True,random_state=42)
score = cross_val_score(GaussianNB(),X,y,cv=skf,scoring="accuracy")
print('Skor setiap fold:',score)
print('Rata-rata score: ',score.mean())
```

```python
#Turning hyperparameter
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=2, shuffle=True,random_state=42)
score = cross_val_score(CategoricalNB(),X,y,cv=skf,scoring="accuracy")
print('Skor setiap fold:',score)
print('Rata-rata score: ',score.mean())
```
